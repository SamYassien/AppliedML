# Fashion MNIST

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow.keras as keras
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.decomposition import PCA
```

Loading data from keras datasets just because it's simpler

```python
fasion_mnist_data = keras.datasets.fashion_mnist
(X_train, y_train), (X_test, y_test) = fasion_mnist_data.load_data()
```

```python
print(f'train images size: {X_train.shape}, train labels size {y_train.shape}')
print(f'test dataset size: {X_test.shape}, test labels size {y_test.shape}')
```

```python
categories = np.unique(y_train)
print(f'number of unique labels: {categories.shape[0]}\n')
print(f'labels {categories}')
```

```python
# label names from https://keras.io/api/datasets/fashion_mnist/
label_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

Display some examples from the dataset

```python
def display_img(image, label):
    plt.grid(False)
    plt.title(label)
    plt.imshow(image)
    plt.xlabel(label)
    plt.axis('off')
```

```python
plt.figure(figsize=(10,10))
for i, (image, label) in enumerate(zip(X_train[:25], y_train[:25])):
    plt.subplot(5,5,i+1)
    display_img(image, label_names[label])
```

```python
# Normalizing the data
x_train = X_train.astype('float32')
x_test = X_test.astype('float32')
x_train /= 255.0
x_test /= 255.0
```

The fashion mnist image size is 28 x 28. When flattened we will have 784 predictors. These are too many parameters and would slow down our training for little gain to accuracy performance. To fix that, we apply dimentionality reduction to the dataset using principle component analysis (PCA) to reduce the parameters the 100 most important predictors to our classification problem.

```python
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1] * x_train.shape[2])
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1] * x_test.shape[2])
print(x_train.shape)
print(x_test.shape)
```

```python
pca = PCA(n_components=100)
pca.fit(x_train)

x_train_pca = pca.transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
print(f'train data shape after PCA: {x_train_pca.shape}')
print(f'test data shape after PCA: {x_test_pca.shape}')
```

**K-nearest neighbors classifier:**

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
knn_accuracy = metrics.accuracy_score(y_test, y_pred_knn)
print(f'Accuracy on test data = {knn_accuracy}')
```

prediction report for each class

```python
print(metrics.classification_report(y_test, y_pred_knn))
```

Confusion Matrix

```python
def plot_conf_matrix(matrix, labels):
    plt.imshow(matrix, interpolation='nearest', cmap=plt.cm.autumn_r)
    plt.title("Confusion Matrix")
    plt.colorbar()
    tick_marks = np.arange(len(labels))
    plt.xticks(tick_marks, labels, rotation=90)
    plt.yticks(tick_marks, labels)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    width, height = matrix.shape
    for x in range(width):
        for y in range(height):
            plt.annotate(str(matrix[x][y]), xy=(y, x), 
            horizontalalignment='center',
            verticalalignment='center')
```

```python
knn_conf_matrix = metrics.confusion_matrix(y_test, y_pred_knn)
plot_conf_matrix(knn_conf_matrix, label_names)
```

**Decision Tree Classifier:**

```python
tree = DecisionTreeClassifier(criterion='entropy')
tree.fit(x_train, y_train)
y_pred_tree = tree.predict(x_test)
tree_accuracy = metrics.accuracy_score(y_test, y_pred_tree)
print(f'Accuracy on test data = {tree_accuracy}')
```

```python
print(metrics.classification_report(y_test, y_pred_tree))
```

```python
tree_conf_matrix = metrics.confusion_matrix(y_test, y_pred_tree)
plot_conf_matrix(tree_conf_matrix, label_names)
```

Random Forrest Classifier

```python
forest = RandomForestClassifier(criterion='entropy')
forest.fit(x_train, y_train)
y_pred_forest = forest.predict(x_test)
forest_accuracy = metrics.accuracy_score(y_test, y_pred_forest)
print(f'Accuracy on test data = {forest_accuracy}')
```

```python
print(metrics.classification_report(y_test, y_pred_forest))
```

```python
forest_conf_matrix = metrics.confusion_matrix(y_test, y_pred_forest)
plot_conf_matrix(forest_conf_matrix, label_names)
```

SGD Classifier

```python
sgd = SGDClassifier()
sgd.fit(x_train, y_train)
y_pred_sgd = sgd.predict(x_test)
sgd_accuracy = metrics.accuracy_score(y_test, y_pred_sgd)
print(f'Accuracy on test data = {sgd_accuracy}')
```

```python
print(metrics.classification_report(y_test, y_pred_sgd))
```

```python
sgd_conf_matrix = metrics.confusion_matrix(y_test, y_pred_sgd)
plot_conf_matrix(sgd_conf_matrix, label_names)
```

RandomForestClassifier() has the highest accuracy at 87%.   
The Confusion matrices show that the most confusion happens between (Pullover & Shirt), (T-shirt/top & Shirt), and (Coat & shirt), which makes since given that they are visually similar.
